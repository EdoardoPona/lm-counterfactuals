```python
def calculated_longest_common_prefix(orig,counter):
    orig = orig.split(" ")
    counter=  counter.split(" ")
    i=0
    for oo, cc in zip(orig, counter):
        if cc != oo:
            return i / len(orig)
            break
        i += 1

    return 1.0

    
    
```

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Increase the default font size
plt.rcParams.update({'font.size': 14})

models = [
    "Meta-Llama-3-8B-Instruct_honest_llama3_8B_instruct_prompt:True.pkl",
    #"gpt2-xl_rome-edits-louvre-rome_prompt:True.pkl",
    #"Llama-2-7b-hf_Llama-2-7b-chat-hf_prompt:True.pkl"
]

names = ["Honest-LLama", "GPT-XL-ROME", "LLama2-Chat"]

name2data = {}
colors = ['blue', 'orange', 'green']  # Define colors for consistency

plt.figure(figsize=(10, 6))

for idx, (name, model) in enumerate(zip(names, models)):
    with open(model, "rb") as f:
        data = pickle.load(f)
        original, counter = data["original"], data["counter"]
        orig = [o.split(" ") for o in original]
        count = [c.split(" ") for c in counter]
        name2data[name] = (orig, count)

        diffs = []
        for o, c in zip(orig, count):
            i = 0
            print(" ".join(o))
            print(" ".join(c))
            print("==================================")
            for oo, cc in zip(o, c):
                if cc != oo:
                    diffs.append(i / len(o))
                    break
                i += 1
        #print(f"{name} diffs: {diffs}")

        # Plot histogram
        plt.hist(
            diffs,
            density=False,
            bins=15,
            alpha=0.5,
            label=name,
            color=colors[idx]
        )

        # Calculate and plot median
        median_diff = np.median(diffs)
        plt.axvline(
            median_diff,
            color=colors[idx],
            linestyle='dashed',
            linewidth=2
        )
        plt.text(
            median_diff,
            plt.ylim()[1]*0.9 - idx*plt.ylim()[1]*0.08,  # Adjust y-position for each label
            f'Median {name}: {median_diff:.2f}',
            rotation=0,
            color=colors[idx],
            verticalalignment='top',
            horizontalalignment='center',
            fontsize=20,  # Increase font size of median labels
            bbox=dict(facecolor='white', alpha=0.5, edgecolor='none')
        )

plt.legend(fontsize=19)
plt.xlabel('Normalized Length of Longest Common Prefix', fontsize=22)
plt.ylabel('Frequency', fontsize=20)
plt.title('Longest Common Prefix Length Distribution', fontsize=24)
#plt.savefig("longest_prefix.pdf", dpi=600)
plt.show()
```
